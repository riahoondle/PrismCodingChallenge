```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.orm import sessionmaker
from sqlalchemy.ext.declarative import declarative_base
import pandas as pd

Base = declarative_base()

class Ttable(Base):
    __tablename__ = 'T data'
    id = Column(Integer, primary_key=True)
    column_a = Column('Message Type', String)
    column_b = Column('Instrument code', String)
    column_c = Column('Exchange Code', String)
    column_d = Column('Security Type', String)
    column_e = Column('Unknown', String)
    column_f = Column('Last price', String)
    column_g = Column('Last volume', String)
    column_h = Column('Bit flag to updated', String)
    column_i = Column('Trade indicator flag', String)
    column_j = Column('Source of realtime message', String)
    column_k = Column('RT of Last Trade in Seconds', Integer)
    column_l = Column('Julian Date of last Trade', String)
    column_m = Column('RT of last update in Seconds', String)
    column_n = Column('Yesterdays close price', String)
    column_o = Column('Open price', String)
    column_p = Column('High price', String)
    column_q = Column('Low pric', String)
    column_r = Column('of trades since market opening', String)
    column_s = Column('Cumulative volume', String)
    column_t = Column('Calculated Turnover', String)
    column_u = Column('VWAP price', String)
    column_v = Column('Origination market of last trade for composite updates', String)
    
class Rtable(Base):
    __tablename__ = 'R data'  # Set the table name
    id = Column(Integer, primary_key=True)
    column_a = Column('Message Type', String)
    column_b = Column('Instrument code', String)

class Stable(Base):
    __tablename__ = 'S data'  # Set the table name
    id = Column(Integer, primary_key=True)
    column_a = Column('Message Type', String)
    column_b = Column('Instrument code', String)
    
class Ctable(Base):
    __tablename__ = 'C data'  # Set the table name
    id = Column(Integer, primary_key=True)
    column_a = Column('Message Type', String)
    column_b = Column('Instrument code', String)
    
class Qtable(Base):
    __tablename__ = 'Q data'  # Set the table name
    id = Column(Integer, primary_key=True)
    column_a = Column('Message Type', String)
    column_b = Column('Instrument code', String)
    
    
# Create a SQLite database engine
engine = create_engine('sqlite:///mini.db')

# Create tables in the database
Base.metadata.create_all(engine)

csv_file = 'Tdata.csv'
df = pd.read_csv(csv_file)

csv_file1 = 'Rdata.csv'
df1 = pd.read_csv(csv_file1)

csv_file2 = 'Sdata.csv'
df2 = pd.read_csv(csv_file2)

csv_file3 = 'Cdata.csv'
df3 = pd.read_csv(csv_file3)

csv_file4 = 'Qdata.csv'
df4 = pd.read_csv(csv_file4)


# Create a session to interact with the database
Session = sessionmaker(bind=engine)
session = Session()

# Insert data into the 'Ttable' table
for _, row in df.iterrows():
    data = Ttable(column_a=row['Message Type'], column_b=row['Instrument code'], column_c=row['Exchange Code'], column_d=row['Security Type'], column_e=row['Unknown'], column_f=row['Last price'], column_g=row['Last volume'], column_h=row['Bit flag to updated'], column_i=row['Trade indicator flag'], column_j=row['Source of realtime message'], column_k=row['RT of Last Trade in Seconds'], column_l=row['Julian Date of last Trade'])
    session.add(data)

# Insert data into the 'Rtable' table
for _, row in df1.iterrows():
    data1 = Rtable(column_a=row['Message Type'], column_b=row['Instrument code'])
    session.add(data1)

# Insert data into the 'Stable' table
for _, row in df2.iterrows():
    data2 = Stable(column_a=row['Message Type'], column_b=row['Instrument code'])
    session.add(data2)
    
# Insert data into the 'Ctable' table
for _, row in df3.iterrows():
    data3 = Ctable(column_a=row['Message Type'], column_b=row['Instrument code'])
    session.add(data3)
    
# Insert data into the 'Qtable' table
for _, row in df4.iterrows():
    data4 = Qtable(column_a=row['Message Type'], column_b=row['Instrument code'])
    session.add(data4) 

session.commit()
```

```python
## Display the data from 'Ttable'
data = session.query(Ttable).all()
for row in data:
    print(row.column_a, row.column_b, row.column_c, row.column_d, row.column_e, row.column_f, row.column_g, row.column_h, row.column_i, row.column_j)

```
